# Probability density function of the cumulative probability function

### Exercise

Simulate a random variable $X$ that follows a standard normal distribution. If the result of the simulation is $x$, calculate with the cumulative probability function the cumulative probability $p(x) = F(x) = P(X \leq x)$. Repeat the simulation $1000$ times, calculate the probability $p$ in each iteration and fill a histogram with its value. Compare the histogram of $p$ with a suitable probability density function.

### Solution

Let $X$ be a continuous random variable following a standard normal distribution:

$$X \sim N(\mu=0, \sigma=1)$$

The cumulative distribution function of X is:

$$F(X=x) = \int_{-\infty}^{x}{\rm d}u\,f(u)$$

Since $X$ is a random variable, $p(x) = F(x)$ is a random variable too. In this Notebook we determine the probability density function of $p$ through Monte Carlo simulations.

%matplotlib inline

```python
# Getting libraries

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, uniform
```

Random number generator

```python
num_seed = 42
rng = np.random.default_rng(num_seed)
```

Monte Carlo simulation

```python
simulations = 1000

x = rng.normal(size=simulations)
p = norm.cdf(x)
```

```python
plt.figure(figsize=(15,8))

plt.subplot(1, 2, 1)
plt.hist(p, bins=10, color='orange', edgecolor='black', linewidth=1.2)
plt.ylim(bottom=0)
plt.xlabel('$p = F(x)$', fontsize=16)
plt.ylabel('Frequency', fontsize=16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

plt.subplot(1, 2, 2)
plt.hist(p, bins=10, density=True)
plt.ylim(bottom=0)
plt.xlabel('$p = F(x)$', fontsize=16)
plt.ylabel('Density', fontsize=16)

x = np.linspace(np.min(p),np.max(p),100)
plt.hist(p, bins=10, color='orange', density=True, edgecolor='black', linewidth=1.2, label='Simulation')
plt.plot(x, uniform().pdf(x), ls = '-', color='blue', lw=5, alpha=0.6, label='PDF U(0,1)')
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.legend(fontsize=16,loc=3)

plt.show()
```

We see above that the CDF follows a uniform distribution between 0 and 1.
